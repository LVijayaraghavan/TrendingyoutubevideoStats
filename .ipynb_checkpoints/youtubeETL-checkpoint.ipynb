```python
# Import all dependencies.
%matplotlib inline
from matplotlib import style
style.use('fivethirtyeight')
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import datetime as dt
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func, distinct,desc,and_
from sqlalchemy.sql.functions import coalesce
from datetime import datetime , timedelta,date
import time
import os
import json

from pandas.io.json import json_normalize
import warnings
warnings.filterwarnings('ignore')
import pymysql
pymysql.install_as_MySQLdb()
from config import sqluser,sqlpassword,dbport,dburi,dbname
```

```python
# Here we are reading the CSV file which contains the videos and the statistics 
file_name=os.path.join("Resources","Usvideos.csv")
youtube_videodf=pd.read_csv(file_name,error_bad_lines=False)
youtube_videodf
```

```python
# Preprocessing .. Here we convert the trending date and the publish date which are in different format to date format
youtube_videodf['trending_date'] = pd.to_datetime(youtube_videodf['trending_date'], format='%y.%d.%m')

youtube_videodf['publish_time'] = pd.to_datetime(youtube_videodf['publish_time']).apply(lambda x : x.date())


#values = {'description':"No Description available"}
#youtube_videodf= youtube_videodf.fillna(value=values)

#checking for any NAN values
youtube_videodf.isna().sum()
```

```python
# Assemble the needed columns in this dataframe
youtube_videodf=youtube_videodf[['video_id','trending_date','title', 'description','category_id','publish_time','views','comment_count','likes','dislikes']]
youtube_videodf
```

```python
# Reading the Json file which contains the Categories for the category ID
filename=os.path.join("Resources",'US_category_id.json')
with open(filename) as f:
    d = json.load(f)
```

```python
# Getting the category from Nested Dict using Json Normalize
items_data = json_normalize(d, 'items',errors='ignore')
#items_data
items_data_final=pd.concat([items_data.drop('snippet', axis=1), pd.DataFrame(items_data['snippet'].tolist())], axis=1)
#items_data_final.to_csv("C:/users/lalit/Beautifulsoup/youtube/ca.csv")

items_data_final= items_data_final.rename(columns={'id':'category_id','title':'category'})
items_data_final['category_id']=items_data_final['category_id'].apply(pd.to_numeric)

items_data_final=items_data_final[['category_id','category']].copy()
items_data_final.set_index('category_id', inplace=True)
items_data_final
```

```python
#Establishing connection to SQL 

engine = create_engine(f"mysql://{sqluser}:{sqlpassword}@{dburi}:{dbport}/{dbname}")
#items_data_final.to_sql(name='youtube_category', con=engine, if_exists='append', index=True)
```

```python
#Perform Pd Merge with the two dataframes so we could get the Category

Youtube_merged_df=pd.merge(youtube_videodf,items_data_final, how='inner', on=['category_id'] )
Youtube_merged_df
```

```python
# Assemling the needed data
Youtube_merged_df=Youtube_merged_df[['video_id','title','trending_date','category','publish_time','views','comment_count'\
                                    ,'likes','dislikes']]
#Check to see if we have an Nulls
Youtube_merged_df.isna().sum()
```

```python
#Rename Columns and write to MYSQL table
Youtube_merged_df=Youtube_merged_df.rename(columns={'publish_time':'publish_date'})
Youtube_merged_df.to_sql(name='youtube_video', con=engine, if_exists='append', index=False)
```

```python
# Group BY Category and aggregate for the total views , likes, dislikes and comments
video_analysis_final=Youtube_merged_df.groupby('category').sum()
```

```python
# Here we are performing more tasks of getting 
video_analysis_final['views_per_comment']=video_analysis_final['views']/video_analysis_final['comment_count']
video_analysis_final['views_per_likes']=video_analysis_final['views']/video_analysis_final['likes']
video_analysis_final['views_per_dislikes']=video_analysis_final['views']/video_analysis_final['dislikes']
video_analysis_final.
```

```python
video_analysis_final.to_sql(name='youtube_summary', con=engine, if_exists='append', index=True)
```
